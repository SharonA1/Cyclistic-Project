Anova tests to cyclistic cleaned_dataset

```r
library("ggplot2")
library("ggpubr")
library("tidyverse")
library("broom")
```

```r
stat_data <- read.csv("/kaggle/input/cyclistic-analysis/cleaned_tripdata.csv")
```

```r
summary(stat_data)
```

# One-way Anova: ride_duraion by meber_casual

```r
oneway_duration <- aov(ride_duration ~ member_casual, data = stat_data)
summary(oneway_duration)
```

# One-way Anova: ride_distance by meber_casual

```r
oneway_distance <- aov(ride_distance ~ member_casual, data = stat_data)
summary(oneway_distance)
```
